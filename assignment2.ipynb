```python
#import funtion
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data
from google.colab import files
uploaded = files.upload()

see_data = pd.read_csv('heart.csv') # Read in csv file to dataframe in Pandas
see_data.head()
```

```python
#split data into  train and test
X = see_data.iloc[:, 0: 12]
X
```

```python
y = see_data.iloc[:, [13]]
y
```

```python
#model training
#split data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=0)
#这里的random_state就是为了保证程序每次运行都分割一样的训练集和测试集。
#否则，同样的算法模型在不同的训练集和测试集上的效果不一样。


X_train
```

```python
y_train
```

```python
#import to model
from sklearn.ensemble import RandomForestClassifier
np.random.seed(1)
model=RandomForestClassifier()

model.fit(X_train,y_train) #use train data training

model.score(X_test,y_test)#
```

```python
#模型预测 (Predict)
y_pred = model.predict(X_test)#
```

```python
#性能指标 (Performance Matrics)

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
# Confusion matrix for the random forest classification
print(confusion_matrix(y_test, y_pred))
```

```python
# Accuracy score
print(accuracy_score(y_test, y_pred))
```

```python
# classification report
print(classification_report(y_test, y_pred))
```
